```python
#TASK 3 - Compare the Triplets
import math
import os
import random
import re
import sys


def compareTriplets(a, b):
    
    result = [1 if x > y
     else -1 
     if y > x 
     else 0 
     for x,y in list(zip(a,b))]
    
    return result.count(1) , result.count(-1)


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
#TASK 5 - Diagonal Difference

import math
import os
import random
import re
import sys




def diagonalDifference(arr):
    
    counter = 0
    counter2 = len(arr) - 1
    first_diag = []
    second_diag = []

    for i in range(len(arr)):
        first_diag.append(arr[counter][counter])
        counter+=1
    print(first_diag)
    counter = 0

    for i in range(len(arr)):
        second_diag.append(arr[counter][counter2])
        counter+=1
        counter2-=1
    print(second_diag)
        

    return abs(sum(first_diag)-sum(second_diag))
   
    
        
        
        
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#TASK 9 - Birthday Cake Candles

import math
import os
import random
import re
import sys


def birthdayCakeCandles(ar):

    ar.sort()

    result = ar.count(ar[len(ar)-1])
    return result


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    ar_count = int(input())

    ar = list(map(int, input().rstrip().split()))

    result = birthdayCakeCandles(ar)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#TASK 10 - Time Conversion

import os
import sys

def timeConversion(s):
    hour, minute, second_AMPM = s.split(':')
     
    if second_AMPM[2:] == "PM" and hour != '12':
        hour = str(int(hour) + 12) 
    if second_AMPM[2:] == "AM" and hour == '12':
        hour = "00" 
    if second_AMPM[2:] == "PM" and hour == '12':
        hour = "12" 
        
    conversion_time = hour + ':' + minute + ':' + second_AMPM[0:2]      
    
    return conversion_time



if __name__ == '__main__':
    f = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    f.write(result + '\n')

    f.close()


```
