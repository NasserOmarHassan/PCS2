```python
from Bio import SeqIO

def pdst(collection):
    for s1 in collection:
        for s2 in collection:
            c = 0
            for i in range(len(s1)):
                if s1[i] != s2[i]:
                    c += 1
            print(c/len(s1), end = ' ')
        print('\n', end = '')
```

```python
collection = []
for record in SeqIO.parse("C://Users//Nasser//Downloads//rosalind_PDST.txt", 'fasta'):
    collection.append(record.seq)
pdst(collection)
```
